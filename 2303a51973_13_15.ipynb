```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample rental data (replace this with your actual data)
data = {
    'rental_date': [
        '2024-11-01', '2024-11-02', '2024-11-03', '2024-11-04', '2024-11-05',
        '2024-11-06', '2024-11-07', '2024-11-08', '2024-11-09', '2024-11-10'
    ],
    'rental_count': [150, 200, 180, 130, 110, 170, 210, 220, 250, 240]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert 'rental_date' to datetime
df['rental_date'] = pd.to_datetime(df['rental_date'])

# Extract the day of the week (0=Monday, 6=Sunday)
df['day_of_week'] = df['rental_date'].dt.dayofweek

# Group by the day of the week and sum the rental counts
day_of_week_rentals = df.groupby('day_of_week')['rental_count'].sum()

# Map day of the week (0=Monday, 6=Sunday) to the actual day name
day_of_week_rentals.index = day_of_week_rentals.index.map({
    0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday',
    4: 'Friday', 5: 'Saturday', 6: 'Sunday'
})

# Sort by the highest rental count
top_rentals_days = day_of_week_rentals.sort_values(ascending=False).head(5)

# Print the top 5 rental days
print("Top 5 days with the most rentals:")
print(top_rentals_days)

# Visualize the data
top_rentals_days.plot(kind='bar', color='skyblue')
plt.title("Top 5 Days with Most Rentals")
plt.xlabel('Day of the Week')
plt.ylabel('Rental Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# Sample bike rental data (replace with your actual dataset)
data = {
    'rental_date': [
        '2024-12-25', '2024-12-31', '2024-11-11', '2024-07-04', '2024-01-01',
        '2024-08-15', '2024-05-30', '2024-10-31', '2024-09-05', '2024-11-25'
    ],
    'rental_count': [500, 700, 450, 800, 650, 400, 350, 300, 500, 550]
}

# Convert to DataFrame
df = pd.DataFrame(data)
df['rental_date'] = pd.to_datetime(df['rental_date'])

# Define a list of holidays (this is a sample, you can modify it or use a holiday API)
holidays = [
    '2024-01-01', '2024-07-04', '2024-12-25', '2024-11-11', '2024-12-31',
    '2024-05-30', '2024-08-15', '2024-10-31', '2024-09-05', '2024-11-25'
]

# Convert the holiday list to datetime
holidays = pd.to_datetime(holidays)

# Add a column to indicate if the date is a holiday
df['is_holiday'] = df['rental_date'].isin(holidays)

# Group by holiday status and sum the rental counts
holiday_rentals = df.groupby('is_holiday')['rental_count'].sum()

# Print the rental count for holidays vs non-holidays
print("Rental Counts on Holidays vs Non-Holidays:")
print(holiday_rentals)

# Visualize the data
holiday_rentals.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Bike Rentals on Holidays vs Non-Holidays')
plt.xlabel('Holiday Status')
plt.ylabel('Total Rental Count')
plt.xticks([0, 1], ['Non-Holiday', 'Holiday'], rotation=0)
plt.show()

# Identify the most popular holidays (those with the highest rental count)
holiday_rentals_sorted = df[df['is_holiday'] == True].sort_values(by='rental_count', ascending=False)
print("\nMost Popular Holidays for Bike Sharing (based on rental count):")
print(holiday_rentals_sorted[['rental_date', 'rental_count']].head(5))

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (replace with your actual dataset)
data = {
    'rental_date': [
        '2024-06-01', '2024-06-02', '2024-06-03', '2024-06-04', '2024-06-05',
        '2024-06-06', '2024-06-07', '2024-06-08', '2024-06-09', '2024-06-10'
    ],
    'rental_count': [150, 200, 180, 220, 250, 270, 190, 210, 230, 250],
    'temperature': [22, 25, 20, 18, 30, 28, 22, 24, 27, 26],  # in Celsius
    'precipitation': [0, 0, 0, 5, 0, 0, 10, 0, 0, 3],  # mm of rain
    'wind_speed': [10, 12, 5, 15, 8, 9, 10, 7, 14, 11]  # in km/h
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a new column to classify weather conditions (for simplicity)
# We will classify based on:
# - Temperature: 20-30C considered ideal for bike rentals.
# - Precipitation: Low precipitation (<= 5mm).
# - Wind Speed: Low to moderate wind speeds (<= 15 km/h).

df['ideal_weather'] = ((df['temperature'] >= 20) & (df['temperature'] <= 30) &
                        (df['precipitation'] <= 5) & (df['wind_speed'] <= 15))

# Calculate the rental counts on ideal weather days
ideal_weather_rentals = df[df['ideal_weather'] == True]['rental_count']

# Calculate the average rental count on ideal weather days
average_rentals_on_ideal_weather = ideal_weather_rentals.mean()

# Print the average rental count on ideal weather days
print(f"Average rentals on days with ideal weather conditions: {average_rentals_on_ideal_weather}")

# Visualize the data
# Plotting rental counts by weather condition
df['weather_condition'] = ['Ideal' if x else 'Non-Ideal' for x in df['ideal_weather']]

# Group by the weather condition and sum the rental counts
weather_rentals = df.groupby('weather_condition')['rental_count'].sum()

# Plot the result
weather_rentals.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Bike Rentals by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Total Rental Count')
plt.xticks(rotation=0)
plt.show()

# Find the weather condition (ideal or non-ideal) with the most rentals
most_suitable_weather = weather_rentals.idxmax()
print(f"The most suitable weather condition for bike rentals is: {most_suitable_weather}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample bike rental data (replace with your actual dataset)
data = {
    'rental_datetime': [
        '2024-11-01 08:30', '2024-11-01 09:15', '2024-11-01 09:45', '2024-11-01 10:00', '2024-11-01 12:30',
        '2024-11-01 14:00', '2024-11-01 16:30', '2024-11-01 18:15', '2024-11-01 19:00', '2024-11-01 21:00'
    ],
    'rental_count': [15, 20, 18, 22, 30, 40, 50, 30, 25, 20]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert rental_datetime to datetime format
df['rental_datetime'] = pd.to_datetime(df['rental_datetime'])

# Extract the hour from the rental_datetime
df['hour'] = df['rental_datetime'].dt.hour

# Group by hour and sum the rental counts
hourly_rentals = df.groupby('hour')['rental_count'].sum()

# Print the hourly rental counts
print("Hourly Rentals:")
print(hourly_rentals)

# Visualize the hourly utilization
hourly_rentals.plot(kind='bar', color='skyblue')
plt.title('Hourly Utilization of Bike Rentals')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Rental Count')
plt.xticks(rotation=0)
plt.show()

# Find the hour with the highest rental count
most_popular_hour = hourly_rentals.idxmax()
print(f"The most popular hour for bike rentals is: {most_popular_hour}:00")
```
